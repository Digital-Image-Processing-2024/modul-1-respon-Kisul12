```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as img
import cv2 as cv
```

Script di atas merupakan sintaks untuk pemanggilan library yang akan digunakan dalam program selanjutnya 

Script di bawah merupakan sintaks yang dimana img.imread() berguna untuk meng-import gambar yang akan digunakan dalam program  gambar dengan library matplotlib.pylot dan ('respon.jpg') merupakan sebuah link gambar yang di import dari device dan code imshow() berfungsi untuk menampikan gambar yang kita import.

```python
image = img.imread('respon.jpg')
plt.imshow(image)
```

script di bawah merupakan sintaks yang berfungsi mengubah draft gambar dari origiginal ke draft gambar RGB yang berasal dari library CV. pada line ke 3 code dibawah berfungsi untuk mengatur saluran warna yang dimana indeks 0, 1, 2 merujuk pada warna Reed, Green, dan Blue. pada code dibawah seluran pada indeks 1(green) danindeks 2(blue) diberi nilai 0 sehingga hanya indeks 0(Reed) saja yang tampak pada gambar yang akan di display

```python
redImage = cv.cvtColor(image, cv.COLOR_BGR2RGB)

redImage[:, :, 1] = 0  # line 3
redImage[:, :, 2] = 0  

# Menampilkan gambar
plt.imshow(redImage)
plt.axis('off')
plt.show()
```

Script di bawah digunakan untuk melakukan iterasi dari 0 sampai -1 dari 256 untuk menghasilkan nilai dari 0-255 yaitu rentang reed level dari soal yang diberikan.dan script bawah pula berfungsi untuk memanggil fungsi bar dari plt untuk membuat sebuah histogram yang berisikan nilai dari reedImage serta menampilkannya

```python
histReed = cv.calcHist([redImage], [0], None, [256], [0, 256])

plt.hist(redImage[:,:,0].ravel(), bins=256, color='red', alpha=0.7)
plt.title('Histogram Gambar Warna Merah')
plt.xlabel('Intensitas Piksel')
plt.ylabel('Frekuensi')
plt.show()
```
